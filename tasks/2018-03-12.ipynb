```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
iris = load_iris()
```

```python
X, y = pd.DataFrame(iris.data[:, 2:], columns=iris.feature_names[2:]), pd.DataFrame(iris.target)
X.head()
```

```python
def plot_samples(marker='o',plot_show=True):
    plt.figure(figsize=[10, 7])
    # First 50 data belong to Iris Sentosa, plotted in green
    plt.plot(X['petal length (cm)'].iloc[:50], X['petal width (cm)'].iloc[:50], marker, color='green', markersize=5, label='Iris Sentosa')
    # Next 50 data belong to Iris Versicolour, plotted in red
    plt.plot(X['petal length (cm)'].iloc[50:100], X['petal width (cm)'].iloc[50:100], marker, color='red', markersize=5,label='Iris Versicolour')
    # Last 50 data belong to Iris Virginica, plotted in blue
    plt.plot(X['petal length (cm)'].iloc[100:150], X['petal width (cm)'].iloc[100:150], marker, color='blue', markersize=5,label='Iris Virginica')

    if plot_show:
        plt.xlim(0,8)
        plt.ylim(-1,3)
        plt.title('Varieties of Iris plants')
        plt.xlabel('Petal length')
        plt.ylabel('Petal width')
        plt.legend(loc='lower right', fontsize='large')
        plt.show()
    
plot_samples()
```

```python
f, ax = plt.subplots(figsize=[10, 7])
sns.kdeplot(X, y, ax=ax, shade=True, shade_lowest=False, cbar=True, legend=True);
sns.rugplot(X['petal length (cm)'], color="black", ax=ax)
sns.rugplot(X['petal width (cm)'], vertical=True, color="black", ax=ax);
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.title('Density estimation using KDE on Iris data');
```
